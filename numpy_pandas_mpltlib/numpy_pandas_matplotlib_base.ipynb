Заполните пропуски пустой строкой для текстовых колонок, и удалите строки с пропусками для колонок с категориальными значениями.

<font color='LightSteelBlue'>**Подсказка**</font>

Значения <font color='CornflowerBlue'>**категориальной**</font> переменной определяют факт принадлежности к какой-то категории, категорий при этом должно быть ограниченное кол-во сильно меньшее размера датасета.

<font color='CornflowerBlue'>**Категориальная переменная**</font> - колонка, которая имеет ограниченное число уникальных значений. Вообще говоря, может быть как типа object (текстовые), так и int (численные). В случае int можно считать, что категории закодированы числами. В таком случае значения нельзя сравнивать между собой и совершать над ними арифметические операции. В этом задании категориальные переменные только типа object.

<font color='OrangeRed'>**Внимание!**</font> Для проверки корректности выведите кол-во пропусков во всех столбцах и общее кол-во строк в таблице - должна быть 3021 запись. Ошибка при удалении строк может привести к неправильному выполнению следующих заданий.




Возможно, пригодятся такие функции:
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html#pandas.DataFrame.fillna
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html#pandas.DataFrame.dropna

```python
# Шаг 1: Удаление строк с пропусками для категориальных колонок
# Мы предполагаем, что категориальной является колонка 'country', так как у нее ограниченное число уникальных значений
data = data[data['country'].notnull()]  # Удаление строк с пропусками в колонке 'country'

# Шаг 2: Заполнение пропусков в текстовых колонках пустыми строками
text_columns = data.select_dtypes(include=['object']).columns  # Получаем все текстовые колонки
data[text_columns] = data[text_columns].fillna('')  # Заполняем пропуски пустыми строками

# Шаг 3: Замена некорректных значений в колонках 'tags' и 'video_id'
data.loc[data['tags'] == '[none]', 'tags'] = ''  # Замена '[none]' на пустую строку
data.loc[data['video_id'] == '#NAME?', 'video_id'] = ''  # Замена '#NAME?' на пустую строку

# Вывод данных для проверки
print(data.isnull().sum())  # Проверка пропусков во всех колонках
print(f"Общее количество строк в таблице: {data.shape[0]}")
```

### <font color='DarkOrange'>**Задание 11 [0.5 баллa]**</font>

Какой видеоролик был опубликован в наибольшем кол-ве стран? В качестве ответа привeдите название видеоролика (колонка title).

<font color='LightSteelBlue'>**Подсказка**</font>

Для решения могут помочь функции [groupby](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html) и [sort_values](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html).

```python
# Группируем данные по видеороликам (video_id) и считаем количество уникальных стран для каждого видео
video_country_count = data.groupby('video_id')['country'].nunique()

# Находим видеоролик с наибольшим количеством стран
most_countries_video_id = video_country_count.idxmax()

# Получаем название видеоролика по его video_id
most_countries_video_title = data.loc[data['video_id'] == most_countries_video_id, 'title'].iloc[0]

# Выводим название видеоролика
print("Видеоролик, опубликованный в наибольшем количестве стран:", most_countries_video_title)
```

Race Highlights | 2022 British Grand Prix

<font color='CornflowerBlue'>**Matplotlib**</font> - основная библиотека визуализации двумерных данных на языке python.

Рисунок создается с помощью последовательного вызова функций из модуля **pyplot** (создание рисунка, задание области рисования, построение графиков в области рисования, украшение графика метками). Основные типы графиков:
- <font color='DarkSeaGreen'>**line plot**</font> — обычный поточечный график, которым можно изображать кривые, соединяющие последовательные точки;
- <font color='DarkSeaGreen'>**histogram**</font> — гистограмма, показывающая распределение некоторое величины;
- <font color='DarkSeaGreen'>**scatter plot**</font> — диаграмма рассеивания, показывающая взаимосвязь двух величин;
- <font color='DarkSeaGreen'>**bar plot**</font> — столбчатая диаграмма, показывающий взаимосвязь количественной величины от категориальной.

### Оформление графиков

Важно, чтобы все графики выглядели опрятно и соответствовали следующим правилам:

🔷 **<font color='plum'>Максимально поясняйте, что изображено на графике</font>**

    ✅ Добавляйте название (plt.title)
    ✅ Подписывайте оси (plt.xlabel, plt.ylabel)
    ✅ Подписывайте единицы измерения на осях
    ✅ Делайте понятные подписи


🔷 **<font color='plum'>Следите за визуальной составляющей графика</font>**

<font color='LightSteelBlue'>**Подсказка**</font>
Не нужно учитывать время публикации, только дату с точностью до дня.

Также обратите внимание, что один и тот же ролик может быть опубликован в нескольких странах, и ему может соответствовать несколько строк в таблице. На графике необходимо учесть каждый ролик только 1 раз.



Сначала может понадобиться перевести колонку даты публикации к типу `datetime`:

```python
data.publishedAt = pd.to_datetime(data.publishedAt)
```

```python
data.publishedAt.dt.date
```

```python
# Убираем дубликаты по video_id
unique_videos = data.drop_duplicates(subset='video_id')

# Преобразуем даты в формат без времени
unique_videos['publishedAt'] = unique_videos['publishedAt'].dt.date

# Считаем количество видеороликов, опубликованных в каждый день
video_count_by_date = unique_videos['publishedAt'].value_counts().sort_index()

# Создаем полный диапазон дат для указанного периода
date_range = pd.date_range(start='2022-06-04', end='2022-07-10').date

# Заполняем пропуски нулями для дней, когда видео не публиковались
full_video_counts = pd.Series(0, index=date_range)
full_video_counts.update(video_count_by_date)

# Строим график
plt.figure(figsize=(10, 6))
plt.plot(full_video_counts.index, full_video_counts.values, marker='.', color='salmon', label='Число роликов')
plt.title('Статистика добавленных видеороликов на YouTube по дате публикации')
plt.xlabel('Дата публикации')
plt.ylabel('Число роликов')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()

# Показываем график
plt.show()
```
