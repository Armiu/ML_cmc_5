```python
data.publishedAt = pd.to_datetime(data.publishedAt)
```

```python
data.publishedAt.dt.date
```

```python
#    video_id
unique_videos = data.drop_duplicates(subset='video_id')

#      
unique_videos['publishedAt'] = unique_videos['publishedAt'].dt.date

#   ,    
video_count_by_date = unique_videos['publishedAt'].value_counts().sort_index()

#       
date_range = pd.date_range(start='2022-06-04', end='2022-07-10').date

#     ,    
full_video_counts = pd.Series(0, index=date_range)
full_video_counts.update(video_count_by_date)

#  
plt.figure(figsize=(10, 6))
plt.plot(full_video_counts.index, full_video_counts.values, marker='.', color='salmon', label=' ')
plt.title('    YouTube   ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()

#  
plt.show()
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
countries = data['country'].unique()
stat = data.groupby('country')
stat = stat['country'].describe()['count']
#      
#   
plt.figure(figsize=(10, 6))
stat.plot(kind='bar', color='hotpink')

#  
plt.title('    ')
plt.xlabel('')
plt.ylabel(' ')
plt.xticks(rotation=45)
plt.grid(axis='y')

#  
plt.tight_layout()
plt.show()
```

### <font color='DarkOrange'>** 3 [1 ]**</font>

  `(scatter plot)`  -     -      (country='FR').          (comments_disabled=True).

    ?

```python


#       
stat = data[(data['country'] == 'FR') & ~data['comments_disabled']]
stat = stat[['view_count', 'comment_count']]

#    
x_vals = [x for x in stat['view_count']]
y_vals = [y for y in stat['comment_count']]

plt.figure(figsize=(16, 9))
plt.suptitle('       ')

plt.xlabel(' ', fontsize=10)
plt.ylabel(' ', fontsize=10)

#   
plt.yscale('log')
plt.xscale('log')

#   
plt.scatter(x_vals, y_vals, s=15, color='rosybrown')

#      
min_val = min(min(x_vals), min(y_vals))
max_val = max(max(x_vals), max(y_vals))
plt.axis([min_val, max_val, min_val, max_val])

#  
plt.grid(True)

#  
plt.show()
```
